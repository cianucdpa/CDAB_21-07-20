## Covid-19

https://documenter.getpostman.com/view/10808728/SzS8rjbc

To pull the data you must first pass the api endpoint into `requests.get()`. This will retrieve the data. Then you parse it using `json()`. Now you can browse it as a dictionary, or convert it to a DataFrame. (Sometimes, I might combine all this into one line of code)

```python
import pandas as pd
```

```python
cv19 = requests.get('https://api.covid19api.com/summary')
cv19 = cv19.json()
pd.DataFrame(cv19['Countries']).head()
```

## Space News

https://www.spaceflightnewsapi.net/documentation#/Article/get_api_v2_articles

```python
space = requests.get('https://spaceflightnewsapi.net/api/v2/articles?_limit=100')
```

```python
pd.DataFrame(space.json()).head()
```

## Universities

https://github.com/Hipo/university-domains-list-api

```python
uni = requests.get('http://universities.hipolabs.com/search?country=ireland')
pd.DataFrame(uni.json()).head()
```

## Weather
https://github.com/robertoduessmann/weather-api

```python
dub = requests.get('https://goweather.herokuapp.com/weather/dublin')
```

```python
pd.DataFrame(dub.json())
```

```python
pd.DataFrame(dub.json()['forecast'])
```

## City Bikes

http://api.citybik.es/v2/

```python
db = requests.get("http://api.citybik.es/v2/networks/dublinbikes").json()
```

```python
db = pd.DataFrame(db)
```

```python
pd.DataFrame(db.loc['stations','network']).head()
```
